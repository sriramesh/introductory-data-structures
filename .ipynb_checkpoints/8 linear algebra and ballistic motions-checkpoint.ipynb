```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
from numpy.linalg import inv
import time
import IPython.display
```

```python
# use linear algebra to chart whether a given ballistic motion (ie ball throw) captured in 25 frames follows a parabolic trajectory
df = pd.read_csv('coordinates.csv', header=None) # load coordinates of the ball throw in each of the 25 frames 
matrix = df.to_numpy() # convert to numpy matrix
```

```python
fig, axs = plt.subplots(5,5, figsize=(20, 20), facecolor='w', edgecolor='k') # display each of the 25 frames into a figure with 5 x 5 subplots 
fig.subplots_adjust(hspace = .5, wspace=.5)
axs = axs.ravel()

for i in range(0, len(df)): # chart the frames in matplotlib
    name = 'frames/frame'+str(i+1)+'.png'
    img = mpimg.imread(name)
    axs[i].imshow(img)
    axs[i].plot(matrix[i][0], matrix[i][1], 'ro', fillstyle='none', lw=2)
```

```python
x, y = [], [] # fit a parabola y = ax^2 + bx + c to the coordinates using least-squares estimation

for i in range(0, len(matrix)):
    x.append(matrix[i][0])
    y.append(matrix[i][1])
    
X = np.stack( (np.square(x), x, np.ones((len(matrix)))), axis=1) # y = ax^2 + bx + c
u = inv(np.transpose(X)@X)@np.transpose(X)@y # use linear algebra to get values of a, b and c and store in vector u

xp = np.arange(0,1400,100) # parabola
yp = u[0]*xp*xp + u[1]*xp + u[2]
```

```python
plt.plot(x,y,'ro', fillstyle='none') # plot the least-squares curve over the 25 coordinates of the ball throw
plt.plot( xp, yp, 'deepskyblue',linestyle='-' ) # parabola
plt.gca().invert_yaxis()
```

```python
# run crude animation of the ball throw, overlaid with the parabola and coordinates.csv data

for i in range(0, len(matrix)): 
    name = 'frames/frame'+str(i+1)+'.png'
    img = mpimg.imread(name)
    plt.figure(figsize=(10,10))
    plt.imshow(img) # show each of the 25 frames
    plt.plot(matrix[i][0],matrix[i][1],'ro', fillstyle='none') # original coordinates.csv data
    plt.plot( xp, yp, 'tab:orange',linestyle='-' ) # parabola in orange
    plt.axis('off')
    plt.show()
    time.sleep(0.3) # delay by 0.3 seconds
    IPython.display.clear_output(wait=True) # wait to clear the output of the current cell receiving output
```
